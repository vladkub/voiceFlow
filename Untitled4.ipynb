```python
# ============================================================
# XTTS-v2 - ФИНАЛЬНАЯ ИСПРАВЛЕННАЯ ВЕРСИЯ
# С понижением версии PyTorch до 2.5.1
# ============================================================

print("🚀 Финальная установка XTTS-v2...")

# 1. Удаляем старые версии PyTorch
!source /content/xtts_env/bin/activate && \
    pip uninstall torch torchaudio torchvision -y

# 2. Устанавливаем PyTorch 2.5.1 (где weights_only=False по умолчанию)
!source /content/xtts_env/bin/activate && \
    pip install torch==2.5.1 torchaudio==2.5.1 --index-url https://download.pytorch.org/whl/cu121

# 3. Устанавливаем TTS
!source /content/xtts_env/bin/activate && \
    pip install TTS==0.21.3

# 4. Создаём скрипт для запуска
script_content = '''
import torch
import gradio as gr
from TTS.api import TTS
import tempfile
import time

print(f"🐍 PyTorch version: {torch.__version__}")
print("🔍 Загрузка модели XTTS-v2...")
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"📱 Устройство: {device}")

# Принудительно отключаем weights_only для загрузки
import warnings
warnings.filterwarnings("ignore")

# Загружаем модель
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2", gpu=(device=="cuda"))

print("✅ Модель загружена!")

def clone_voice(audio_file, text, language="ru", speed=1.0):
    if audio_file is None:
        return None, "❌ Загрузите образец голоса!"
    if not text.strip():
        return None, "❌ Введите текст!"

    try:
        start = time.time()
        output_path = tempfile.NamedTemporaryFile(delete=False, suffix=".wav").name

        tts.tts_to_file(
            text=text,
            speaker_wav=audio_file,
            language=language,
            file_path=output_path,
            speed=speed
        )

        elapsed = time.time() - start
        return output_path, f"✅ Готово за {elapsed:.1f} сек"
    except Exception as e:
        return None, f"❌ {str(e)[:150]}"

with gr.Blocks(title="XTTS-v2 Voice Cloner") as demo:
    gr.Markdown("""
    # 🎤 XTTS-v2 Voice Cloner

    **Клонирование голоса в Google Colab**

    1. Загрузите образец голоса (5-10 секунд)
    2. Введите текст
    3. Нажмите "Сгенерировать"
    """)

    with gr.Row():
        with gr.Column():
            audio_input = gr.Audio(type="filepath", label="🎙️ Образец голоса")
            text_input = gr.Textbox(label="📝 Текст", lines=3, value="Привет! Это тест клонирования голоса.")
            lang = gr.Dropdown(["ru", "en", "es", "fr", "de", "it", "pt", "pl", "zh-cn"], label="Язык", value="ru")
            speed = gr.Slider(0.5, 2.0, 1.0, label="Скорость")
            btn = gr.Button("🔊 Сгенерировать", variant="primary")
        with gr.Column():
            audio_out = gr.Audio(label="🎵 Результат")
            status = gr.Textbox(label="Статус", lines=2)

    btn.click(clone_voice, [audio_input, text_input, lang, speed], [audio_out, status])

demo.launch(share=True)
'''

with open('/content/run_xtts_final.py', 'w') as f:
    f.write(script_content)

print("✅ Скрипт создан!")
print("\n🚀 ЗАПУСК...")
!source /content/xtts_env/bin/activate && python /content/run_xtts_final.py
```
